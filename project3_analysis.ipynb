Project 3: ChIP-seq Analysis

**INTRODUCTION**  

RUNX1 is a transcription factor which plays a major role in the development of cancer due to its significant role in hematopoiesis. In breast cancer cells, RUNX1 is known to control gene expression, interact with chromatin-remodeling machinery by tethering to Erα (estrogen receptor). However, binding of RUNX1 to the DNA and regulating the specific genes in breast cancer is still unclear. To address this, the authors in this study combined several genome-wide bioinformatic approaches like ChIP-seq to map RUNX1 binding sites across the genome, RNA-seq to measure how RUNX1 depletion alters gene expression, and Hi-C to capture the 3D structure of chromatin and the associated domains. By integrating these datasets computationally, they could understand the correlation of RUNX1 to transcriptional changes and the 3D organization of the genome, and test for RUNX1 as a factor shaping chromatin interactions in breast cancer cells. 



**METHODS**

**Read Quality Assessment**  
Initial quality assessment of the raw read FASTQ files was evaluated using FastQC (v0.12.1) based on various metrics like per-base quality, adapter contamination, GC content, and sequence duplication levels. A MultiQC report was generated to compile all the quality control metrics across various samples, helping detect any trends or biases. Default parameters were set for both the processes.

**Adapter Trimming, Genome Indexing and Alignment**   
Low-quality bases and Illumina sequencing adapters were removed using Trimmomatic (v0.39). Both trimmed reads and Trimmomatic log files were collected as the outputs. Reads were then aligned to the human reference genome GRCh38 using Bowtie2 (v2.5.3) and a Bowtie2 index was generated using the reference FASTA file. The trimmed paired-end reads were aligned with default parameters to produce SAM files, which were then converted to the BAM format.

**Sorting, Indexing, and Alignment Statistics**  
All BAM files were sorted and indexed using Samtools (v1.22.1) and the alignment quality was assessed using samtools flagstat. It reported the proportion of mapped reads and other alignment metrics. Results from FastQC, Trimmomatic, and Flagstat were aggregated into the MultiQC report summarizing quality across all samples.

**Generation of bigWig Coverage Tracks**  
To quantify read coverage across the genome, the sorted and indexed BAM files were processed using deeptools bamCoverage (v3.5.6). The bigWig files were generated using default bin sizes and normalization settings to provide the per-base coverage tracks which were used in the downstream correlation and signal profiling analyses.

**Correlation Analysis of bigWig Signals**  
Sample similarity was evaluated using deeptools multiBigwigSummary, which produced a coverage matrix for all samples. Correlation heatmaps were generated using deeptools plotCorrelation with Pearson correlation to assess reproducibility between replicates and distinguish between the samples and the controls.

**Peak Calling and generation of reproducible peaks with HOMER**  
Peak calling was performed using HOMER (v4.11) where each BAM file was converted into a HOMER tag directory. Peaks were then identified using findPeaks for each replicate and its corresponding input. Replicate specific peak call files were produced which were later converted to BED format. Peaks from replicate 1 and replicate 2 were intersected using bedtools intersect (v2.31.1) and the peak which overlapped by at least 1 bp in both the replicates were retained. 

**Filtering the blacklisted regions and peak annotation**  
Signal-artifact regions caused due to false enrichment were removed using the ENCODE and any peaks overlapping a blacklisted region were discarded. This helped produce a clean, and reproducible peak set. The filtered reproducible peaks were then annotated based on the GRCh38 reference FASTA and the Ensembl GTF annotations using HOMER, providing information about the promotor, enhancer, distance to transcription start site. 
                                                                           
**Gene Body Signal Profiling**  
BED file containing the TSS and TES of all hg38 genes was obtained using the UCSC Table Browser. Deeptools was used to quantify RUNX1 ChIP-seq signal across the gene bodies around 2000 bp flanking regions and matrices were generated for the samples.

**Motif Enrichment Analysis**  
RUNX1 binding motifs were identified using HOMER and enrichment was calculated relative to random genomic background regions. The results showed known motif enrichments, de novo motif predictions, sequence logos along with significance scores. 

**Integration with RNA-seq data and functional analysis**  
The differential expression results from the publication were downloaded from the GEO study described in the original paper. Based on the significant thresholds in the study, RUNX1 ChIP-seq peaks were overlapped with differentially expressed genes to recreate Figure 2F from the paper. Figure 2D and 2E by using the IGV genome browser to visualize the RUNX1 binding at two highlighted genes. Functional enrichment was performed using the annotated peak set through the GREAT enrichment method to visualise the enriched molecular pathways and regulatory programs associated with RUNX1 binding.

**QUALITY CONTROL EVALUATION**

The MultiQC report showed that each sample had millions of mapped reads. INPUT_rep2 had the lowest read count with 10.0 million mapped reads, while INPUT_rep1 and IP_rep2 had the highest counts with 28.1 million mapped reads. FASTQC investigated a number of parameters and indicated that most of the basic statistics pass for all the samples. There were some issues with the per base sequence content and per sequence GC content. For the per base sequence content, the INPUT data failed and there was a warning for the IP samples. This occurs because CHIP-seq enriches only a certain part of the genome and not the entire genome, resulting in a bias. High duplication levels were seen in the IP samples, with 89.1% duplication in IP_rep1 and 74.3% in IP_rep2. This is consistent with successful enrichment at binding sites, which is also supported by the presence of overrepresented sequences for 2.1% in IP_rep1 and 3.13% in IP_rep2. The adapter-content module passed for all four samples, showing no significant adapter contamination. The Trimmomatic results suggest that only a small fraction of the reads were trimmed, indicating that the samples were nearly clean. The alignment statistics showed strong performance with high mapping rate across all the samples.  

Overall, the high read depth, minimal adapter contamination, and strong alignment rates indicate that the data was of high quality and suitable for downstream ChIP-seq analysis. The observed GC bias and high duplication rates were expected artifacts of ChIP-seq experiments. To further improve data quality, reducing PCR amplification cycles during library preparation and optimizing fragmentation could help minimize duplication while maintaining strong peak enrichment.

**MOTIF FINDING**

The HOMER-known motif finding results showed that the top enriched motif is a RUNX family motif, with a significantly high p-value of 1e-610. It is seen to be present in 28.62% of the target sequences with 5.31% background, suggesting that the RUNX motifs appear nearly six times more frequently. These results demonstrate a strong enrichment, and help recover the expected motif with high confidence and are highly consistent with the expectations of the ChIp-seq experiment. Since RUNX proteins recognize similar DNA motifs and overlapping gene elements, the ChIP-seq method was successful in capturing specifically the RUNX regions and omitting the background noise.

**ANALYSIS OF ANNOTATED PEAKS USING GREAT** 
 
 The GREAT enrichment analysis showed that the genomic regions associated with the peaks were enriched for processes related to placental development, embryonic development, and blood vessel formation. These pathways were consistent with the role of RUNX1 as a transcription factor involved in hematopoietic and other development related networks. The region–gene association plot indicates that mainly all the peaks are located near Transcription Start Site (TSS), within a few kilobases of the promoter regions. This arrangement showed the correlation between the RUNX1 binding regions and the function of transcriptional activation and gene regulation, which is consistent with the peak signal distribution observed in the ChIP-seq data.
